OUTLIER ANALYSIS

```python
#using iqr method
#using box plot
import numpy as np
df1=df.select_dtypes(include=np.number)
```

```python
for i in df1:
    q1=df1[i].quantile(0.25)
    q2=df1[i].quantile(0.50)
    q3=df1[i].quantile(0.75)
```

```python
IQR=q3-q1
```

```python
IQR
```

```python
lw=q1-1.5*IQR
uw=q3+1.5*IQR
```

```python
outliers = ((df1[i]<lw)&(df1[i]>lw))
```

```python
print(outliers.sum())
```
